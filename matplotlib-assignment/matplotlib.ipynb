```python
import matplotlib.pyplot as plt
import numpy as np
```

# Exercise 1

```python
# Create two arrays: x, y  and z where:
#    x contains 12 evenly spaced numbers between 1 and 20 
#    y contains the values of x multiplied by 2 ie. 2*x
#    z contains the squares of elements of x ie x**2
x = np.linspace(1, 20, 12)
y = x * 2
z = x ** 2
```

```python
# Plot y vs x, set the label of x-axis as ‘x’, the label of y-axis as ‘y’ and the title of the plot as ‘y = 2x’. Increase the x axis range to (0, 25) and y axis range to (0, 450)

plt.xlabel('x')
plt.ylabel('y')
plt.title('y = 2x')

plt.plot(x, y)
plt.xlim(0, 25)
plt.ylim(0, 450)
plt.show()
```

```python
# Create a new plot and plot y vs x and z vs x in the same plot. Differentiate between the two  functions using legend.
plt.plot(x, y, label = "y = 2x")
plt.plot(x, z, label= "z = x^2")
plt.legend()
plt.show()

```

```python
# Create a figure containing two subplots. Set the size of the figure to be 12X6. In the first subplot, plot y vs x and in the second subplot, plot z vs y. Set the title and axis labels accordingly.

fig, axes = plt.subplots(1, 2, figsize =  (10, 5))

axes[0].plot(x, y)
axes[0].set_title("y = 2x")
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x, z)
axes[1].set_title("z = x^2")
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')

plt.show()
```

```python
import pandas as pd

df = pd.read_csv('pima_diabetes.csv')
df
```

```python
#  Plot a bar plot of the ‘outcome’ variable. Is the dataset balanced or unbalanced? Try to explore the consequences of having such a dataset.

sns.countplot(df.Outcome).set(title= "Barplot Outcome")
```

The dataset is unbalanced since number of zeros is double of number of ones.

The consequences of having unbalanced dataset is that when we train our model using such data set the model tends to accurately predict the outcome with higher number but fails to predict the outcome with less number

```python
 # Create a figure containing 8 subplots arranged in (2, 4) grid. Plot the histograms of the 8 features (all the columns except the outcome) present in the dataset. Identify the features having normal distribution and skewed distributions.
fig, axes = plt.subplots(2, 4,  figsize=(20, 15))

for i in range(0, 2):
    for j in range(0, 4):
        sns.histplot(data=df.iloc[:, i*4+j],ax=axes[i][j])
```

Glucose, Blood Pressure and BMI seems to have a normal distribution while other have skewed distribution

```python
# Create a figure containing 8 subplots arranged in (4, 2) grid. Plot the boxplot for each of the features in each of the subplots. Outliers are a set of observations that vary significantly from the other observations. Which feature do you think has the most outliers? And Why?
fig, axes = plt.subplots(4, 2,  figsize=(10, 15))

for i in range(0, 4):
    for j in range(0, 2):
        sns.boxplot(data=df.iloc[:, i*2+j],ax = axes[i][j]).set(title=df.columns[i*2+j])
```

It seems that Insulin and DiabetesPedigreeFunction have the most number of outliers because as seen in boxplot these two have most number of values outside maximum whisker points.

```python
# A correlation matrix represents how much the features are correlated to each other. You can compute the correlation matrix of a dataframe using the dataframe.corr() method. Use matplotlib.pyplot.matshow to plot the correlation matrix in matplotlib. Also, you can use the matplotlib.pyplot.colorbar() to better explain your correlation matrix.

corr_matrix = df.corr()

corr_plt = plt.matshow(corr_matrix)
plt.colorbar(corr_plt)

plt.show()
```
