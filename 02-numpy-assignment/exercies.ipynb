```python
np.sum(mat, axis = 0)       #sum of array columnwise
np.sum(mat, axis = 1)       #sum of array rowwise
np.amax(mat, axis = 0)      #max element columnwise
np.amax(mat, axis = 1)      #max element rowwise
np.mean(mat, axis = 0)      #average of array columnwise
np.mean(mat, axis = 1)      #average of array rowwise

mat.astype('float64')
mat.astype('object')
```

```python
def normalization(mat):
    range = np.max(mat) - np.min(mat)
    return (mat - np.min(mat)) / range
 
 
def standardization(mat):
    mu = np.mean(mat, axis=0)
    sigma = np.std(mat, axis=0)
    return (mat - mu) / sigma

print(normalization(mat))
print(standardization(mat))
```

# Exercise 9

```python
def convolve_1d(array, kernel):
    kernel_shape = kernel.shape[0]
    output_shape = array.shape[0] - kernel_shape + 1
    conv_array = np.zeros(output_shape, dtype= int)
    for row in range(output_shape):
        conv_array[row] = (array[row:row+kernel_shape]*kernel).sum()
    return conv_array

given_array = np.arange(10) 
convolved_array = convolve_1d(given_array, np.array([1,2,3]))
print(convolved_array)
```

# Exercise 10

```python
def convolve_2d(image, kernel):

    output = np.zeros_like(image)
    
    image_padded = np.zeros((image.shape[0] + (kernel.shape[0]-1), 
                             image.shape[1] + (kernel.shape[1]-1)))   
    image_padded[(kernel.shape[0]//2):-(kernel.shape[0]//2), 
                 (kernel.shape[1]//2):-(kernel.shape[1]//2)] = image
    
    for col in range(image.shape[1]):
        for row in range(image.shape[0]):
            output[row,col]=(kernel*image_padded[row:row + kernel.shape[0], col:col + kernel.shape[1]]).sum()
    return output

image = cv2.imread("/home/araul/Desktop/ML_Intern/numpy-assignment/me.png") 
image = cv2.cvtColor(src=image, code=cv2.COLOR_BGR2GRAY) 
image.shape
kernel = np.ones((3,3))
kernel.shape[0]
convolved_image = convolve_2d(image,kernel)
plt.imshow(convolved_image)
```
